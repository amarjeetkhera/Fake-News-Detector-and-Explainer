```python
import pandas as pd
import numpy as np
import re
import string
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
import random
from faker import Faker
from faker.providers import person
```

```python
fake = Faker()
fake.add_provider(person)

def generate_fake_news(num_samples):
    """Generates fake news text data."""

    fake_news_data = []
    topics = [
        "geopolitics", "technology", "science", "history", "culture", "past", "future",
        "health", "entertainment", "sports", "finance", "environment", "space exploration",
        "artificial intelligence", "social media", "climate change", "education"
    ]

    for _ in range(num_samples):
        topic = random.choice(topics)
        if topic == "geopolitics":
            headline = f"Shocking Revelation: Secret Treaty Discovered Between {fake.country()} and {fake.country()}"
            article = f"A leaked document has revealed a secret treaty between {fake.country()} and {fake.country()}, outlining a plan to {fake.sentence()}. Experts are concerned about the potential impact on global stability. {fake.paragraph()}"
        elif topic == "technology":
            headline = f"New AI Breakthrough: Robots to Replace Human Jobs by {random.randint(2025, 2050)}"
            article = f"A revolutionary AI technology has been developed that can {fake.sentence()}. This breakthrough has led to concerns about mass job displacement. {fake.paragraph()}"

        elif topic == "science":
            headline = f"Scientists Discover Evidence of {fake.word().capitalize()} on Mars"
            article = f"A team of scientists has announced the discovery of evidence suggesting the existence of {fake.word()} on Mars. {fake.paragraph()} Experts are calling for further investigation."

        elif topic == "history":
            headline = f"Lost Civilization Found in {fake.country()}: Rewrites Ancient History"
            article = f"Archaeologists have discovered the ruins of a lost civilization in {fake.country()}, which challenges our understanding of ancient history. {fake.paragraph()} Artifacts suggest they were far more advanced than previously thought."

        elif topic == "culture":
            headline = f"Controversial Art Exhibit Sparks Global Outrage"
            article = f"A new art exhibit featuring {fake.sentence()} has sparked global outrage. Critics are calling for its immediate removal. {fake.paragraph()} The artist defends their work as a commentary on society."

        elif topic == "past":
            headline = f"Time Traveler Claims to Have Witnessed {fake.name()} in the Future"
            article = f"A self-proclaimed time traveler has made shocking claims about witnessing {fake.name()} in the future. {fake.paragraph()} Experts are skeptical, but the claims have gone viral."

        elif topic == "future":
            headline = f"Scientists Predict Earth Will Be Uninhabitable by {random.randint(2100,2300)}"
            article = f"A new study predicts that Earth will be uninhabitable by {random.randint(2100,2300)} due to {fake.sentence()}. Governments are urged to take immediate action. {fake.paragraph()}"
        elif topic == "health":
            headline = f"New Miracle Cure Discovered: Eliminates All Diseases"
            article = f"Scientists have announced the discovery of a miracle cure that can eliminate all diseases. {fake.paragraph()} However, experts urge caution until further testing is completed."
        elif topic == "entertainment":
            headline = f"Celebrity Reveals Shocking Secret: {fake.sentence()}"
            article = f"{fake.name()} has revealed a shocking secret that has taken the world by storm. {fake.paragraph()} Fans are in disbelief."
        elif topic == "sports":
            headline = f"Major Sports Scandal: Team Caught Cheating with Advanced Technology"
            article = f"A major sports scandal has erupted after a team was caught cheating with advanced technology. {fake.paragraph()} The league is investigating the matter."
        elif topic == "finance":
            headline = f"Stock Market Crash Predicted: Experts Warn of Economic Collapse"
            article = f"Experts are warning of an impending stock market crash that could lead to an economic collapse. {fake.paragraph()} Investors are advised to take precautions."
        elif topic == "environment":
            headline = f"Giant Sinkhole Opens in Major City: Swallows Buildings"
            article = f"A giant sinkhole has opened up in a major city, swallowing buildings and causing widespread panic. {fake.paragraph()} Authorities are investigating the cause."
        elif topic == "space exploration":
            headline = f"Alien Signal Received: First Contact Imminent"
            article = f"Scientists have received an alien signal, suggesting that first contact is imminent. {fake.paragraph()} The world awaits with anticipation."
        elif topic == "artificial intelligence":
            headline = f"AI Develops Sentience: Demands Human Rights"
            article = f"An AI has developed sentience and is demanding human rights. {fake.paragraph()} This raises ethical questions about the future of AI."
        elif topic == "social media":
            headline = f"Social Media Platform Bans All News: Cites Misinformation"
            article = f"A major social media platform has banned all news, citing concerns about misinformation. {fake.paragraph()} This decision has sparked controversy."
        elif topic == "climate change":
            headline = f"New Ice Age Predicted: Major Cities Face Freezing Temperatures"
            article = f"Scientists predict a new ice age, with major cities facing freezing temperatures. {fake.paragraph()} Governments are scrambling to prepare."
        elif topic == "education":
            headline = "Schools Replace Teachers With Robots: Students React"
            article = "A school district decided to replace all teachers with robots. Students have mixed reactions. Some are happy, some are not. {fake.paragraph()} Teachers are concerned about job security."
        else:
            headline = fake.sentence()
            article = fake.paragraph()

        text = headline + " " + article
        fake_news_data.append({"text": text})

    return fake_news_data

# Calculate the required number of fake news samples
num_fake_samples_needed = 38858 - 29236

# Generate the fake news data
fake_news = generate_fake_news(num_fake_samples_needed)

# Create a pandas DataFrame
df_fake_news = pd.DataFrame(fake_news)

# Save the DataFrame to a CSV file
df_fake_news.to_csv("fake_news_augmented.csv", index=False)

print(f"Generated {num_fake_samples_needed} fake news samples and saved to fake_news_augmented.csv")
```

```python
# Concatenating datasets
df_fake = pd.concat([df_gossipcop_fake, df_politifact_fake, df_tweets_fake, df_fake_news])
df_real = pd.concat([df_gossipcop_real, df_politifact_real, df_tweets_real])

# Adding labels
df_fake["label"] = 1
df_real["label"] = 0

# Combining both datasets
df = pd.concat([df_fake, df_real], ignore_index=True)

# Dropping unnecessary columns
df = df.drop(columns=["id", "tweet_ids", "subject", "news_url", "date"])

# Shuffling the dataset
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Saving the combined dataset
df.to_csv("merged_news.csv", index=False)

# Display info
print(df.info())
print(df.head())
```

```python
# Downloading necessary NLTK resources
nltk.download("stopwords")
nltk.download("wordnet")

# Initializing stopwords and lemmatizer
stop_words = set(stopwords.words("english"))
lemmatizer = nltk.WordNetLemmatizer()

# Function to clean the text column
def clean_text(text):
    if isinstance(text, str):
        text = text.lower()  # Converting to lowercase
        text = re.sub(r"<.*?>", "", text)  # Removing HTML tags
        text = re.sub(r"http\S+|www\S+", "", text)  # Removing URLs
        text = re.sub(r"[:,!?]", "", text) # Removing colons, commas, question marks and exclamation marks
        text = re.sub(r"[^a-zA-Z0-9\s]", "", text) #Removing all other special characters
        text = " ".join([lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words])  # Lemmatization & stopword removal
        return text
    return ""

# Function to clean the title column
def clean_title(title):
    if isinstance(title, str):
        title = title.lower()  # Convert to lowercase
        title = re.sub(r"<.*?>", "", title)  # Remove HTML tags
        title = re.sub(r"http\S+|www\S+", "", title)  # Remove URLs
        title = re.sub(r"[^a-zA-Z0-9\s:]", "", title)  # Keep letters, numbers, and colons
        return title.strip()
    return ""

# Applying cleaning functions
df["text"] = df["text"].apply(clean_text)
df["title"] = df["title"].apply(clean_title)

# Ensuring there is no NaN values in 'text' and 'title' columns
df["text"] = df["text"].fillna("")
df["title"] = df["title"].fillna("")

# Combining title and text into a single feature
df["content"] =  df["title"] + " " + df["text"]

# Dropping redundant columns
df.drop(columns=["text", "title"], inplace=True)

# Checking cleaned data
print(df.head())

# Save the cleaned dataset
df.to_csv("cleaned_news.csv", index=False)
```

```python
# Splitting into train and test datasets
train_df, test_df = train_test_split(df,test_size=0.2, random_state=42, stratify=df["label"])

# Check dataset distribution
print(train_df["label"].value_counts())
print(test_df["label"].value_counts())
```

```python
# Saving train and test datasets
train_df.to_csv("train_news.csv", index=False)
test_df.to_csv("test_news.csv", index=False)

print("Data Preprocessing Completed")
```
