```python
import pandas as pd
```

```python
df_gossipcop_real = pd.read_csv("/content/drive/MyDrive/Fake News/Datasets/gossipcop_real.csv")
df_gossipcop_fake = pd.read_csv("/content/drive/MyDrive/Fake News/Datasets/gossipcop_fake.csv")
df_politifact_real = pd.read_csv("/content/drive/MyDrive/Fake News/Datasets/politifact_real.csv")
df_politifact_fake = pd.read_csv("/content/drive/MyDrive/Fake News/Datasets/politifact_fake.csv")
df_tweet_real = pd.read_csv("/content/drive/MyDrive/Fake News/Datasets/True.csv")
df_tweet_fake = pd.read_csv("/content/drive/MyDrive/Fake News/Datasets/Fake.csv")
```

```python
# Concatenating real and fake datasets
df_real = pd.concat([df_gossipcop_real, df_politifact_real, df_tweet_real])
df_fake = pd.concat([df_gossipcop_fake, df_politifact_fake, df_tweet_fake])
```

```python
# Adding target labels
df_real["label"] = 0
df_fake["label"] = 1
```

```python
# Combining both datasets
df = pd.concat([df_real, df_fake], ignore_index=True)

# Dropping unnecessary columns
df = df.drop(columns=["id", "news_url", "tweet_ids", "subject", "date"])

# Ensuring there is no NaN values
df["text"] = df["text"].fillna("")
df["title"] = df["title"].fillna("")

# Shuffling the dataset
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Saving the combined dataset
df.to_csv("merged_news.csv", index=False)

# Display info
print(df.info())
print(df.head())
```
